```python
sales_products_fill=df.fillna({
    "region":df["region"].mode()[0],
    "prodcut_category":df["product_category"].mode()[0]
})
sales_products_fill
```

```python
df["sales_amount"].fillna(df["sales_amount"].median())
```

```python
df["quantity"].fillna(method="ffill")
```

```python
df["customer_age"].fillna(df["customer_age"].mean().round())
```
