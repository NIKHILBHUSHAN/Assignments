```python
import sqlite3
conn=sqlite3.connect("bookstore.db")
cursor=conn.cursor()
cursor.execute("""
   CREATE TABLE IF NOT EXISTS books (
      book_id INTEGER PRIMARY KEY AUTOINCREMENT,
      title TEXT NOT NULL,
      author TEXT NOT NULL,
      price REAL NOT NULL,
      stock_quantity INTEGER DEFAULT 0)"""
)
cursor.execute("""
  CREATE TABLE IF NOT EXISTS customers (
      customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
      name TEXT NOT NULL,
      email TEXT NOT NULL,
      city TEXT,
      join_date TEXT)"""
)
cursor.execute("""
CREATE TABLE IF NOT EXISTS orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    book_id INTEGER,
    quantity INTEGER NOT NULL,
    order_date TEXT NOT NULL,
    total_amount REAL NOT NULL,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY (book_id) REFERENCES books(book_id)
    )
""")
```

```python
cursor.execute("PRAGMA table_info(books)")
books_table_info=cursor.fetchall()
for column in books_table_info:
  print(column)
```

```python
cursor.execute("PRAGMA table_info(customers)")
customers_table_info=cursor.fetchall()
for column in customers_table_info:
  print(column)
```

```python
cursor.execute("PRAGMA table_info(orders)")
orders_table_info=cursor.fetchall()
for column in orders_table_info:
  print(column)
```

```python
import pandas as pd
books=pd.read_sql("SELECT * FROM BOOKS",conn)
customers=pd.read_sql("SELECT * FROM CUSTOMERS",conn)
orders=pd.read_sql("SELECT * FROM ORDERS",conn)
```

```python
print(books.head(3))
print(customers.head(3))
print(orders.head(3))
```

```python
report=pd.merge(orders,customers,on='customer_id')
finalreport=pd.merge(report,books,on='book_id')
finalreport[['order_id','name','city','title','quantity','total_amount']]

```

```python
discounts_data = {
      'book_id': [1, 2, 3, 4, 5],
       'discount_percent': [10, 15, 5, 20, 12]
}
discounts=pd.DataFrame(discounts_data)
discounts.to_sql('discounts',conn,if_exists='replace',index=False)


```

```python
cursor.execute("select * from discounts")
row=cursor.fetchall()
for rows in row:
  print(rows)
```

```python
cursor.execute("""
SELECT books.title,
       books.price,
       discounts.discount_percent,
       books.price - (books.price * discounts.discount_percent / 100) AS discounted_price
       FROM books
       INNER JOIN discounts
       ON books.book_id = discounts.book_id
 """)

rows = cursor.fetchall()

for row in rows:
      print(row)
conn.close()
```
